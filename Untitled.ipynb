```python
song=train_set.drop(["pop","top genre","artist","title"],axis=1)

song_labels=train_set["pop"].copy()
```

creating pipeline

```python
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
my_pipeline=Pipeline([('std_scaler',StandardScaler())])
```

```python
song_tr=my_pipeline.fit_transform(song)
song_tr.shape
# song.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
# model=LinearRegression()
# model=DecisionTreeRegressor()
model=RandomForestRegressor()
model.fit(song_tr,song_labels)
```

```python
some_data=song.iloc[:5]
some_labels=song_labels.iloc[:5]
prepared_data=my_pipeline.transform(some_data)
```

```python
model.predict(prepared_data)
```

```python
list(some_labels)
```

evaluating the model

```python
import numpy as np
from sklearn.metrics import mean_squared_error
song_predictions=model.predict(song_tr)
mse=mean_squared_error(song_labels,song_predictions)
rmse=np.sqrt(mse)
rmse
```

Cross Validation

```python
from sklearn.model_selection import cross_val_score
scores=cross_val_score(model,song_tr,song_labels,scoring="neg_mean_squared_error",cv=10)
rmse_scores=np.sqrt(-scores)
```

```python
rmse
```

```python
def print_scores(scores):
    print("Scores: ",scores)
    print("Mean: ",scores.mean())
    print("Standard deviation: ",scores.std())
```

```python
print_scores(rmse_scores)
```

```python
X_test=test_set.drop(["pop","title","artist","top genre"],axis=1)
Y_test=test_set["pop"].copy()
X_test_prepared=my_pipeline.transform(X_test)
final_predictions=model.predict(X_test_prepared)
final_mse=mean_squared_error(Y_test,final_predictions)
final_rmse=np.sqrt(final_mse)
final_rmse
```
